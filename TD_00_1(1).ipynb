## V. Les instructions conditionnelles
Une instruction conditionnelle est une instruction que ne s'éxecute que si une certaine condition est réalisée. Cette condition s'écrit le plus souvent sous la forme d'un **test de comparaison**. En général, l'instruction conditionnelle n'est exécutée que si le résultat de ce test est vrai.

### V.1 Les opérateurs de comparaison
 Soient x et y deux **variables numériques**, alors on peut tester :
 - leur **égalité** avec l'opérateur `x == y`
 - leur **non-égalité** avec l'opérateur `x != y`
 - le **supérieur ou égal** avec l'opérateur `x >= y`
 - le **plus petit ou égal** avec l'opérateur `x <= y`
 - la **stricte supériorité** avec l'opérateur `x > y`
 - la **stricte infériorité** avec l'opérateur `x < y`
 
Dans tous les cas, ces tests ne peuvent valoir que True ou False ; ce sont des **booléens**.

*Remarques :*
- Les tests d'égalité et de non égalité marchent aussi avec des chaîne de caractères ou des booléens.
- Ils sont par contre à éviter avec des nombres flottants (nombres à virgules) car les erreurs d'arrondis peuvent rendrent différents des nombres qui devraient être égaux (voir IV.1).

### V.2 La structure conditionnelle
En python, une structure conditionnelle s'écrit en toute généralité de la façon suivante :
> ``` python
if test == True :
    instruction1
else :
    instruction2
```

Cela signifie que `instruction1` sera exécutée si le résultat du test est vrai et que sinon, c'est l'`instruction2` qui sera exécutée.

Comme la véracité du test est la condition par défaut pour exécuter la première instruction, le code précédent est complétement équivalent au code suivant :
> ``` python
if test :
    instruction1
else :
    instruction2
```

**<span style="color:red;">IMPORTANT :</span>** Remarquez la présence du caractère ":" à la fin des lignes commençant par `if` et `else`, ainsi que le décalage, appelé **indentation**, des lignes qui suivent ce caractère. Cela peut sembler anodin mais sans cela, le programme ne fonctionnera pas, ou pire, il fonctionnera mal !

*Remarques :*
- La présence du `else` n'est pas obligatoire.
- On peut tout à fait exécuter l'`instruction1` si le résultat du test est faux en mettant comme condition `test == False` ou `test != True` (la première expression est conseillée).
- On peut aussi combiner plusieurs tests simultanément avec les opérateurs logiques `and`, `or`et `not` ("et", "ou" et "non"). Par exemple, l'expression `(test1 and test2) == true` n'est vraie que si test1 et test2 sont vrais simultanément.

## VI. Les structures itératives (boucles)
Une structure itérative permet de **répéter une instruction ou un bloc d'instructions un nombre entier de fois**. Ce nombre de répétition peut être fixé à l'avance (boucle `for`) ou bien il peut dépendre de la réalisation d'une condition (boucle `while`)

### VI.1 La boucle for
En python, les façons les plus courantes d'écrire une boucle `for` sont :
> ```python
\# i varie de 0 à 9 par pas de 1 ==> 10 exécutions des instructions
for i in range(10):
    instructions
```

> ```python
\# i varie de 1 à 9 par pas de 1 ==> 9 exécutions des instructions
for i in range(1,10):
    instructions
```

> ```python
\# i varie de 0 à 9 par pas de 2 ==> 5 exécutions des instructions
for i in range(0,10,2):
    instructions
```

> ```python
\# i varie de 10 à 1 par pas de -1 ==> 10 exécutions des instructions
for i in range(10,0,-1):
    instructions
```

*Remarques :*
- Comme pour les structures conditionnelles, la ligne qui commence par `for` finit par un ":" et la suivante est indentée.
- En python, les lignes commençant par le caractère "#" ne sont pas interprétées, ce sont des **lignes de commentaires**.
- Cette syntaxe utilise la fonction `range(debut, fin, pas)` qui renvoie en réalité une liste d'entier dont la première valeur est égale à `debut` et les suivantes sont incrémentées de `pas` en s'arrêtant juste avant d'atteindre ou de dépasser `fin`.
- Par définition, une boucle for ne peut pas être une boucle infinie mais par contre, elle peut très bien ne jamais exécuter les instructions qu'elle contient.



- On peut parcourir une chaine de caractères avec une boucle for avec le code suivant :
>``` python
\# déclaration de la variable maphrase
maphrase='Ceci est une chaîne de caractères.'
\# pour chaque valeur de l'indice i dans l’intervalle [0,len(maphrase)-1]
for i in range(len(maphrase)):
    print(maphrase[i])
```

- Ou, mieux encore, avec :
        
>``` python
\# déclaration de la variable maphrase
maphrase='Ceci est une chaîne de caractères.'
\# pour chaque(for) caractère(car) dans(in) maphrase
for car in maphrase:
    print(car)
```

### VI.2 La boucle while
On peut par exemple écrire une boucle while de la façon suivante :
> ``` python
i=0
pas=1
while i<10 :
    instructions
    i=i+pas
```

Tant que la condition i < 10 est vérifiée, les instructions contenues dans la boucle while sont exécutées. 

Il faut faire très attention avec une boucle while car si la condition reste toujours vraie, on crée une **boucle infinie** de laquelle le programme ne pourra jamais sortir. Pour éviter cette boucle infinie, il faut s'assurer que les instructions contenues dans la boucle while sont à même de rendre fausse la condition de sortie de la boucle.

Ici, i part de 0 et est incrémenté d'un `pas` qui vaut 1 à chaque tour de boucle. Cela continue tant que i est strictement inférieur à 10, la boucle est donc exécutée de i = 0 à i = 9, soit 10 fois. Cette boucle while est finalement équivalente à une boucle for i in range(10).

- #### Le jeu du nombre mystère
Le principe est simple, l’ordinateur choisit aléatoirement un entier compris entre 1 et 100, et l’utilisateur doit le trouver.  
Nous aurons besoin de la bibliothèque `random` pour utiliser la fonction `randint(a,b)` qui renvoie un nombre entier aléatoire entre a et b compris.
Dans cet algorithme, on initialise une variable `nombre_a_tester` à 0, puis, tant que le nombre n'est pas trouvé, on invite l’utilisateur à entrer un nombre et on lui indique si le nombre cherché est plus grand ou plus petit.

Voilà le programme:
> ```python
from random import *
nombre_a_trouver = randint(1,100)
nombre_a_tester = 0
while nombre_a_tester != nombre_a_trouver:
    nombre_a_tester = int(input("veuillez entrer un entier entre 1 et 100 : "))
    if nombre_a_tester > nombre_a_trouver :
        print("C’est moins !")
    if nombre_a_tester < nombre_a_trouver :
        print("C’est plus !")
print("Bravo ! Vous avez trouvé :",nombre_a_tester)
```

## VII. Les fonctions
Une fonction est une sorte de raccourci vers un morceau de code qui a une tâche bien précise

### VII.1 Déclaration d'une fonction

En général, la fonction prend des **variables en entrée**, on les appelle les **paramètres** de la fonction et **renvoie un résultat**. On parle souvent de **procédure** pour les fonctions qui **ne renvoient pas de résultat** (elle modifie seulement les paramètres ou elle affiche quelque chose par exemple).

L'intérêt d'une fonction est à la fois de **rendre plus lisible un programme** en le découpant en tâches simples réalisées par des fonctions aux noms explicites et de le simplifier en **évitant de recopier plusieurs fois des morceaux de codes presques identiques**.

Une fonction peut être sauvegardée dans un fichier séparé du programme principal et appelée par ce dernier.

En python, une fonction est définie par l'instruction suivante :  
`def mafonction(parametre1, parametre2, ...):`

C'est l'instruction `return` qui permet de préciser ce que renvoie la fonction.

Quelques exemples de fonction :
>``` python
\# fonction avec 1 paramètre et un retour
def f(x) :
    return x**2+1
\# on appelle la fonction par son nom, ici 5 fois
for i in range(5) :
    print(f(i))
```

>``` python
\# fonction sans paramètres et sans retour
def mafonction() :
    for i in range(5) :
    print(i*’*’) \# On appelle la fonction
mafonction()
```

>``` python
\# fonction avec 2 paramètres et un retour
def max(a,b) :
    if a >= b :
        return a
    else :
        return b
print(max(25,13))
```

### VII.2 Spécification d'une fonction
Pour bien faire, toute fonction se doit d’être **spécifiée**, c'est-à-dire d'être livrée avec son **mode d'emploi**. Pour cela, il suffit d’écrire en début de fonction le type des paramètres d’entrées et les conditions sur
ces paramètres, ce que fait la fonction et enfin le type de retour; le tout entre des triples guillemets.

Par exemple pour notre fonction `max(a,b)` précédente cela pourrait donner :
>```python
def max(a,b):
""" a est un entier, b est un entier. Compare les valeurs de a et b et renvoie le plus grand des 2."""
if a >= b :
    return a
else :
    return b
```
